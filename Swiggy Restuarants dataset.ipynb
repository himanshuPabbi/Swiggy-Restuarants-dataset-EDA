```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("swiggy.csv")
```

```python
df
```

```python
df=df.drop("ID",axis=1)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.boxplot(df['Price'])
```

```python
q1=df['Price'].quantile(0.25)
```

```python
q3=df['Price'].quantile(0.75)
```

```python
IQR=q3-q1
```

```python
lower=q1-1.5*IQR
```

```python
upper=q3+1.5*IQR
```

```python
df=df[(df['Price']>lower) & (df['Price']<upper)]
```

```python
df
```

```python
sns.boxplot(df['Avg ratings'])
```

```python
sns.boxplot(df['Delivery time'])
```

```python
q1=df['Delivery time'].quantile(0.25)
```

```python
q3=df['Delivery time'].quantile(0.75)
```

```python
IQR=q3-q1
```

```python
lower=q1-1.5*IQR
```

```python
upper=q3+1.5*IQR
```

```python
df=df[(df['Delivery time']>lower) & (df['Delivery time']<upper)]
```

```python
df
```

```python
sns.boxplot(df['Total ratings'])
```

```python
q1=df['Total ratings'].quantile(0.25)
```

```python
q3=df['Total ratings'].quantile(0.75)
```

```python
IQR=q3-q1
```

```python
lower=q1-1.5*IQR
```

```python
upper=q3+1.5*IQR
```

```python
df=df[(df['Total ratings']>lower) & (df['Total ratings']<upper)]
```

```python
df
```

# Highest Rating Resturant in Bangalore

```python
City_Bangalore_df=df[df['City']=="Bangalore"]
```

```python
City_Bangalore_df
```

```python
highest_rating_df=City_Bangalore_df[['Restaurant','Avg ratings']]
```

```python
highest_rating_df
```

```python
top_rating_resturant_df=highest_rating_df.groupby("Restaurant")['Avg ratings'].mean().sort_values(ascending=False)[:10]

```

```python
top_rating_resturant_df
```

```python
sns.barplot(x=top_rating_resturant_df.values,y=top_rating_resturant_df.index)
plt.show()
```
